```python
import pandas as pd
medical_dataset=pd.read_csv(r'C:\Users\Osama\.jupyter\data.csv') 
medical_dataset
```

```python
set(medical_dataset['diseases'])
```

```python
from sklearn.model_selection import train_test_split
train_data, test_data = train_test_split(medical_dataset, test_size=0.003, random_state=42)
print(test_data)
print(train_data.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier(random_state=1)
train_diagnoses=train_data['diseases']
train_patient_symptoms=train_data.drop(columns=['diseases'])
model.fit(train_patient_symptoms, train_diagnoses)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
test_patient_symptoms=test_data.drop(columns=['diseases'])
test_diagnoses=test_data['diseases']
pred=model.predict(test_patient_symptoms)
print(pred)
```
